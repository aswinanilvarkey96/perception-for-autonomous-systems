```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cma
```

```python
cap = cv2.VideoCapture('Challenge.mp4')
```

```python
ret, frame = cap.read()
```

```python
sift = cv2.xfeatures2d.SIFT_create()
bf = cv2.BFMatcher()
ret, frame1 = cap.read()
ret, frame2 = cap.read()
b,g,r = cv2.split(frame1) # Changing the order from bgr to rgb so that matplotlib can show it
img1 = cv2.merge([r,g,b])
b,g,r = cv2.split(frame2) # Changing the order from bgr to rgb so that matplotlib can show it
img2 = cv2.merge([r,g,b])
img3 = img2.copy()

gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

kp, des = sift.detectAndCompute(gray1, None) #Keypoints and descriptors
kp2, des2 = sift.detectAndCompute(gray2, None)
matches = bf.match(des, des2)

matches = sorted(matches, key = lambda x:x.distance)
img3 = cv2.drawMatches(gray1,kp,gray2,kp2,matches[:10],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

kp_img = cv2.drawKeypoints(gray1, kp, img1, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
cv2.imshow('image', kp_img)
cv2.waitKey(20)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
bf = cv2.BFMatcher()
while True:
    ret, frame1 = cap.read()
    ret, frame2 = cap.read()
    if ret == False:
        break
    
    b,g,r = cv2.split(frame1) # Changing the order from bgr to rgb so that matplotlib can show it
    img1 = cv2.merge([r,g,b])
    b,g,r = cv2.split(frame2) # Changing the order from bgr to rgb so that matplotlib can show it
    img2 = cv2.merge([r,g,b])
    img3 = img2.copy()

    gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
    gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
    
    kp, des = sift.detectAndCompute(gray1, None) #Keypoints and descriptors
    kp2, des2 = sift.detectAndCompute(gray2, None)
    matches = bf.match(des, des2)
    
    matches = sorted(matches, key = lambda x:x.distance)
    img3 = cv2.drawMatches(gray1,kp,gray2,kp2,matches[:100],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

        
    cv2.imshow('image', img3)
    cv2.waitKey(20)
```
